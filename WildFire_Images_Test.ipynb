```python
import subprocess
import sys
from kaggle.api.kaggle_api_extended import KaggleApi

def install(pkg: str):
    """Install a Python package via pip."""
    subprocess.check_call([sys.executable, "-m", "pip", "install", pkg])

# 1) Install required packages
for pkg in [
    "kaggle",          # Kaggle CLI & API
    "torch",           # if you need PyTorch
    "torchvision",     # for ImageFolder, transforms, etc.
    "torchmetrics",    # your metrics lib
    "tqdm",            # progress bars
]:
    install(pkg)

# 2) Download & unzip the Wildfire dataset
api = KaggleApi()
api.authenticate()  # needs ~/.kaggle/kaggle.json or env vars set

dataset_slug = "elmadafri/the-wildfire-dataset"
target_dir   = "data/wildfire"

# This will create `data/wildfire/` and unzip all files into it
api.dataset_download_files(
    dataset_slug,
    path=target_dir,
    unzip=True,
    quiet=False  # show progress
)

print(f" Dataset downloaded & extracted to '{target_dir}'")
```
